```python
import numpy as np;
import matplotlib.pyplot as plt;
import matplotlib.animation as animation;

%matplotlib nbagg


###################################

def f(x):
    return x*x + 2*x - 1;

def tan_f (x, point, slope):
    return slope * (x-point) + f(point);

# def derivative (fun,x):
#     h = 5e-10;
#     return (fun (x+h) - fun (x-h)) / (2*h);

def derivative_f (x):
    return 2*x + 2;

###################################

fig = plt.figure ();
ims = [];

xs = np.linspace (-1,7);
ys = f(xs);


x = 5;
epsilon = 1e-10;


while (np.abs (f(x)) > epsilon):
    y = f(x);
    d = derivative_f (x);
    
    vy = np.linspace (0,y);
    vx = x*np.ones_like (vx);
    
    im = plt.plot (xs, ys, 'b', xs, tan_f(xs, x, d), 'r', vx, vy, 'b--', xs, np.zeros_like (xs), '#808080');
    ax = plt.gca ();
    
    # ax.set_xticks([np.sqrt(2)-1, x]);
    # ax.set_xticks(range (-1,8,1), minor=True);
    # ax.set_yticks([0]);
    # ax.set_yticks(range (-10,70,10), minor=True);
    # ax.grid (which='major');
    
    ims.append (im);  
    x = x - y/d;
    print ('{} {}'.format (x, f(x)));



y = f(x);
d = derivative_f (x);

vy = np.linspace (0,y);
vx = x*np.ones_like (vx);

im = plt.plot (xs, ys, 'b', xs, tan_f(xs, x, d), 'r', vx, vy, 'b--', xs, np.zeros_like (xs), '#808080');
ax = plt.gca ();
ims.append (im);
   
    
ani = animation.ArtistAnimation(fig, ims, interval=500)
plt.show()

```
